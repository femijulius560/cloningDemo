```python
from IPython.display import Markdown, display

def rainbow_text(text):
    colors = ['red', 'orange', 'gold', 'green', 'lightblue', 'yellow', 'violet']
    result = ' '.join(
        f'<span style="color:{c}; font-weight:bold;">{ch}</span>'
        for ch, c in zip(text, colors * (len(text)//len(colors)+1))
    )
    display(Markdown(result))

rainbow_text("Well Done! You've Cloned This Repo!")
```

```python
import matplotlib.pyplot as plt
import numpy as np

theta = np.linspace(0, 6*np.pi, 1000)
r = theta**0.5
x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure(figsize=(6,6))
plt.scatter(x, y, c=theta, cmap='plasma', s=5)
plt.axis('off')
plt.title("âœ¨ Repo Clone Spiral âœ¨", fontsize=16, fontweight='bold')
plt.show()
```
